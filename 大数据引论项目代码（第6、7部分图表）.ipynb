```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
models = ['Lasso Regression', 'Decision Tree', 'Random Forest', 'SVC', 'KNN', 'MLP']
attrition_recall = [55.0769, 83.6923, 84.3077, 58.4615, 49.8462, 83.3846]
new_attrition_recall = [84.3077, 90.1538, 94.7692, 85.2308, 83.0769, 89.5385]

# 设置图形大小和样式
plt.figure(figsize=(12, 8))
plt.style.use('seaborn-darkgrid')

# 绘制柱状图
bar_width = 0.4
index = np.arange(len(models))
bar1 = plt.bar(index, attrition_recall, bar_width, label='Attrition Recall', color='#F39C12' , linewidth=1.5)
bar2 = plt.bar(index + bar_width, new_attrition_recall, bar_width, label='New Attrition Recall', color='#2E86C1',  linewidth=1.5)

# 添加数值标签
def add_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black', fontsize=10)

add_labels(bar1)
add_labels(bar2)

# 添加标签、标题和图例
plt.xlabel('Models', fontsize=14)
plt.ylabel('Recall (%)', fontsize=14)
plt.title('Attrition Recall and New Attrition Recall for Different Models', fontsize=16)
plt.xticks(index + bar_width / 2, models, fontsize=12)
plt.legend(fontsize=12)

# 去掉上边框和右边框
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# 显示柱状图
plt.show()

```

```python
import matplotlib.pyplot as plt

# 数据
models = ['Lasso Regression', 'Decision Tree', 'Random Forest', 'SVC', 'KNN', 'MLP']
precision = [75.2101, 82.9268, 92.5676, 85.2018, 76.7773, 84.1615]
recall = [55.0769, 83.6923, 84.3077, 58.4615, 49.8462, 83.3846]
f1_score = [0.635879, 0.833078, 0.882448, 0.693430, 0.604478, 0.837712]

# 绘制柱状图
bar_width = 0.2
index = range(len(models))

fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(index, precision, bar_width, label='Precision')
bar2 = ax.bar([i + bar_width for i in index], recall, bar_width, label='Recall')
bar3 = ax.bar([i + 2 * bar_width for i in index], f1_score, bar_width, label='F1 Score')

ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Performance Metrics by Model')
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(models)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
models = ['Lasso Regression', 'Decision Tree', 'Random Forest', 'SVC', 'KNN', 'MLP']
precision = [75.2101, 82.9268, 92.5676, 85.2018, 76.7773, 84.1615]
recall = [55.0769, 83.6923, 84.3077, 58.4615, 49.8462, 83.3846]
f1_score = [0.635879, 0.833078, 0.882448, 0.693430, 0.604478, 0.837712]

# 绘制柱状图
bar_width = 0.2
index = range(len(models))

fig, ax1 = plt.subplots(figsize=(15, 6))

bar1 = ax1.bar(index, precision, bar_width, label='Precision', color='#2E86C1')
bar2 = ax1.bar([i + bar_width for i in index], recall, bar_width, label='Recall', color='g')

# 创建第二个坐标轴
ax2 = ax1.twinx()
bar3 = ax2.bar([i + 2 * bar_width for i in index], f1_score, bar_width, label='F1 Score', color='orange', alpha=0.5)

# 设置坐标轴标签
ax1.set_xlabel('Models')
ax1.set_ylabel('Precision and Recall', color='black')
ax2.set_ylabel('F1 Score', color='black')

# 设置柱状图标签
ax1.set_xticks([i + bar_width for i in index])
ax1.set_xticklabels(models)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 将F1-score纵坐标轴放缩到0-1之间
ax2.set_ylim(0, 1)

plt.title('Performance Metrics by Model')
plt.show()
```
